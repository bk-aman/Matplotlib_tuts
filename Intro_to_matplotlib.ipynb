##  Load Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic Graphs

```python
x = [1,2,3]
y = [2,4,6,]
# adding keywords for plots
# plt.plot(x,y, label = '2x', color = 'b', linewidth = 2, marker = '.', markersize = 10, linestyle = '--', markeredgecolor='red' )

#setting figure size
plt.figure(figsize = (5,3), dpi = 100)

#shorthand notation
#fmt = '[color][marker][line]'

#plot basic graphs
plt.plot(x,y, 'b-', label = '2x')

#making series from 0, 4.5
x2 = np.arange(0, 4.5, 0.5)
# print(x2)

#plotting things in parts
plt.plot(x2[:3], x2[:3]**2,'r', label = 'x**2')
plt.plot(x2[2:], x2[2:]**2,'r--')

#add title
plt.title('Our First graph!', fontdict = {'fontname':'Comic Sans MS', 'fontsize':20})
#add label
plt.xlabel('x-axis')
plt.ylabel('y-label')

#add ticks for x and y axis i.e scale of graph
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10,12,14,16])

#add legends
plt.legend()

#save with 300 dpi to get good resolution
plt.savefig('myfirstgraph.png', dpi = 300)
plt.show()
```

## Bar Charts

```python
labels = ['A','B','C']
values = [1,4,2]
bars = plt.bar(labels, values)

pattern = ['/','*','o']

for bar in bars:
    bar.set_hatch(pattern.pop())

# bars[0].set_hatch('/')
# bars[1].set_hatch('*')
# bars[2].set_hatch('O')


plt.show()
```

### Fifa data loaing

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head()
```

### Histograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins, color='grey')

plt.xticks(bins)
plt.xlabel("Rating of Player")
plt.ylabel("Number of Player")

plt.title('Distribution of player as per rating in Fifa 2018')

plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['left', 'right']
colours = ['blue', 'orange']

plt.title('Foot preferrece of FIFA 18 players')
plt.pie([left, right], labels = labels, colors = colours, autopct = '%.2f%%')
```

```python
# fifa.Weight
fifa['Weight'] = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa['Weight'] < 125].count()[0]
light_medium =fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] < 150)].count()[0]
medium = fifa.loc[(fifa['Weight'] >= 150) & (fifa['Weight'] < 175)].count()[0]
medium_heavy = fifa.loc[(fifa['Weight'] >= 175) & (fifa['Weight'] < 200)].count()[0]
heavy = fifa.loc[fifa['Weight'] >= 200].count()[0]

weights = [light, light_medium, medium,medium_heavy, heavy]
labels = ['light', 'light_medium', 'medium', 'medium_heavy', 'heavy']

plt.title('Weight Distribution of Fifa 2018 players ')
plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = [0.4,0.4,0.1,0.0,0.8])
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona'].Overall
madrid = fifa.loc[fifa.Club == 'Real Madrid'].Overall
revs = fifa.loc[fifa.Club == 'New England Revolution'].Overall
Manc = fifa.loc[fifa.Club == 'Manchester City'].Overall

plt.figure(figsize = (5,8))
plt.style.use('default')

plt.title('Professional Soccer Team Comparision')
label = ['FCB', 'RMA', 'NER', 'Manc']
plt.ylabel('Rating of Player')
plt.boxplot([barcelona, madrid, revs, Manc], labels =label)
plt.show()
```
